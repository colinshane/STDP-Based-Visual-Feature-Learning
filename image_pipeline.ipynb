# Image Pipeline

Convert images from RGB to greyscale for processing

* Prep environment

```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

* Image resizing 
(from https://stackoverflow.com/questions/44650888/resize-an-image-without-distortion-opencv)

```python
def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized
```

* Load a single image (TODO: Make this read the directory...)

```python
img = cv2.imread('images/IMG_0827.jpg',cv2.IMREAD_GRAYSCALE)
img_x = img.shape[1]
img_y = img.shape[0]
print('Image dimensions: {}x{}'.format(img_x, img_y))
```

* Resize image to a more appropriate size

```python
img_resized = image_resize(img, width=400)
img_x = img_resized.shape[1]
img_y = img_resized.shape[0]
print('Image dimensions: {}x{}'.format(img_x, img_y))
```

* Crop the face

```python
face = img_resized[150:500, 50:350]
```

* Display what we have

```python
plt.axis("off")
plt.imshow(face, cmap="gray")
plt.show()
```

* Now save the image out

```python
cv2.imwrite('images/IMG_0827_gray.jpg', face)
```
